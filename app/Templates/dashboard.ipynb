```python
import dash
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import templates
import inspect
import pandas as pd
import numpy as np
```

## Create DF

```python
classes = []
standards = []
topics = []
grade_levels = []
for name, obj in inspect.getmembers(templates):
    if inspect.isclass(obj):
        classes.append(obj)
        
classes
```

```python
for j in range(len(classes)):
    x = classes[j]()
    print(x)
    standard = x.attr1 
    topic = x.attr2 
    topics.append(topic)
    grade_level = x.attr3 
    standards.append(standard)
    grade_levels.append(grade_level)
```

```python
df = pd.DataFrame({"Class": classes, "Grade_level": grade_levels, "Standard": standards, "Topic": topics})
df.to_csv("4th grade dataframe")
df
```

```python
topics = list(df['Topic'])
topics
```

## Create Dashboard

```python
markdown_text = '''
Please use this dashboard to select the number and type of SOL questions you would like to practice. Once you select the questions, they will display directly on the webpage.

Please insert yhour answers directly in the field provided. Good luck!'''
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
```

```python
app = JupyterDash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div(
    [
        html.H1("SOL Math Dashboard"),    
        
        dcc.Markdown(children = markdown_text),
        
        html.Div([
            
            html.H3("Grade"),

            dcc.RadioItems(id='grade_level',
                options=[4],
                value=4),

            html.H3("Standard"),

            dcc.RadioItems(id='standard',
                options=['Computation and Estimation','Number and Number Sense', 'Measurement and Geometry', 'Probability, Statistics, Patterns, Functions, and Algebra'],
                value='Computation and Estimation'),

            html.H3("Topic"),

            dcc.RadioItems(id='topic',
                options=[{'label': i, 'value': i} for i in topics]),
            
            html.H3("Number of Questions"),

            dcc.RadioItems(id='num_questions',
                options=[{'label': i, 'value': i} for i in range(25)])

        ], style={'width': '25%', 'float': 'left'})
    ]
)

#figure out what componenet ID and component property will be here
@app.callback(Output(component_id="graph",component_property="figure"), 
                  [Input(component_id='grade_level',component_property="value"),
                   Input(component_id='standard',component_property="value"),
                   Input(component_id='topic',component_property="value"),
                  Input(component_id='num_questions',component_property="value")])

def pull_question_with_topic(grade_level, standard, topic, num_questions):
    import warnings
    warnings.filterwarnings('ignore')
    temp_df= df.query("Standard == '{standard}' & Topic == '{topic}' & Grade_level == {grade_level}".format(standard=standard, grade_level=grade_level, topic=topic))
    temp_df = temp_df.sample(n = num_questions, replace= True)
    for i in range(num_questions):
        target_class = temp_df.iloc[i]['Class']
        x= target_class()
        print(x.print_question())
        try:
            print(x.print_choices())
            student_answer = input()
            print(x.student_answer(student_answer))
        except:
            student_answer = input()
            if "Fraction Addition/Subtraction" in x.attr2:
                try:
                    student_answer= student_answer.split("/")
                    num = int(student_answer[0])
                    denom = int(student_answer[1])
                    student_answer = num/denom
                    print(x.student_answer(student_answer))
                except:
                    print("Next time, please enter your answer as a fraction such as 1/8.")
            if "Fraction Ordering" in x.attr2:
                try:
                    fraction = student_answer.split(", ")
                    numbers= []
                    for item in fraction:
                        item = item.split("/")
                        numbers.append(item)
                    numbers2= []
                    for number in numbers:
                        num = int(number[0])
                        denom= int(number[1])
                        answer = num/denom
                        numbers2.append(answer)
                    student_answer = numbers2
                    print(x.student_answer(student_answer))
                except:
                    print("Next time, please enter your answer as a list of fractions separated by commas and a space.")
            else:
                print(x.student_answer(student_answer))
            
if __name__ == '__main__':
    app.run_server(mode='inline', host = "0.0.0.0",  debug=True, port=8050)
```
